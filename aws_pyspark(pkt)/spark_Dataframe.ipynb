```python
from pyspark.sql.functions import udf,lit,col 
from pyspark.sql.types import IntegerType

def total_sal(salary,bonus):
    return salary+bonus 

total_salary = udf(lambda x,y: total_sal(x,y),IntegerType())
```

```python
employee = spark.read.options(inferschema=True).csv('data/OfficeData.csv',header=True)
```

```python
employee.withColumn('Total_sal',total_salary(employee.salary,employee.bonus)).show()
```

```python
employee.withColumn('Total_sal',total_sal(employee.salary,employee.bonus)).show()
```

```python
def increment(state:str,salary:int,bonus:int):
    if state=="NY":
        salary = salary/10 
        bonus = 0.05*bonus 
    if state=="CA":
        salary = 0.12*salary
        bonus = 0.03*bonus 
    return int(salary+bonus )

incre_sal = udf(increment,IntegerType())
```

```python
employee.withColumn('increment',incre_sal(col('state'),col('salary'),col('bonus'))).show()
```

```python
increment('NY',90000,10000)
```
